### 회귀 모델 만들기

#### 자전거 대여 예측하기

불필요한 컬럼 일단 제거

```python
df = df.drop(['casual', 'registered'], axis = 1)
```

```python
df.to_csv('data/bike_rentals_cleaned.csv', index = False)
```

피처, 타겟 준비

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

모델 라이브러리 import 

```python
import xgboost as xgb
xgb.set_config(verbosity = 0)

from sklearn.linear_model import LinearRegression
```

#### 선형회귀 모델

```python
lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
```

평가

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
rmse = np.sqrt(mse)
```

```python
print('RMSE: %0.2f'%(rmse))
```

결과 분석 : RMSE가 좋은 것인가?
    
RMSE로 성능을 평가하려면 타겟 데이터를 확인해봐야 한다.

```python
df['cnt'].describe()
```

#### XGBRegressor

```python
from xgboost import XGBRegressor
```

### 교차 검증

```python
from sklearn.model_selection import cross_val_score
```

선형회귀 교차검증

```python
model = LinearRegression()

scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv = 10)
```

```python
scores
```

```python
rmse = np.sqrt(-scores)
```

```python
rmse
```

```python
# 평균 제곱근 오차를 출력합니다.
print('회귀 rmse:', np.round(rmse, 2))

# 평균 점수를 출력합니다.
print('RMSE 평균: %0.2f' % (rmse.mean()))
```

```python
# 같은 결과

-np.mean(cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=10))
```

crossvalidate : cross_val_score는 cross_validate 함수를 랩핑한 것이다. crossvalidate를 사용하면 여러개의 지표를 동시에 사용할 수도 있다.

```python
from sklearn.model_selection import cross_validate
```

```python
cv_results = cross_validate(model, X, y , scoring='neg_root_mean_squared_error', cv=10)
```

```python
cv_results
```

- 더 많은 결과를 보여준다.

```python
-np.mean(cv_results['test_score'])
```

XGBoost 교차검증

```python
model = XGBRegressor(objective='reg:squarederror')
```

```python
model
```

```python
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv = 10)
```

```python
rmse = np.sqrt(-scores)
```

```python
rmse
```

```python
# 평균 제곱근 오차를 출력합니다.
print('회귀 rmse:', np.round(rmse, 2))

# 평균 점수를 출력합니다.
print('RMSE 평균: %0.2f' % (rmse.mean()))
```

피처와 타깃 분리

```python
#'income_ <=50K' 열을 삭제합니다.
df = df.drop('income_ <=50K', axis=1)
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

#### 로지스틱 회귀

```python
def cross_val(classifier, num_splits = 10):
    model = classifier
    scores = cross_val_score(model, X, y, cv=num_splits)
    
    # 정확도 출력
    print('정확도:', np.round(scores, 2))

    # 평균 정확도 출력
    print('평균 정확도: %0.2f' % (scores.mean()))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# LogisticRegression으로 cross_val 함수를 호출합니다.
cross_val(LogisticRegression())
```

#### XGBClassifier

```python
from xgboost import XGBClassifier
```

```python
cross_val(XGBClassifier(n_estimators = 5))
```
